```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from pandas_datareader import data as wb
from datetime import date
import datetime
from datetime import datetime, timedelta
import plotly
import plotly.graph_objs as go
```

```python
current_date = date.today()
new_date = current_date - timedelta(days=3739)
mcard_index = wb.DataReader('MA', data_source='yahoo', start=(new_date))['Adj Close']
names = ['MasterCard', 'VISA','American Express', 'Nasdaq']
```

```python
# Since March 2011

tickers = ['MA', 'V', 'AXP','^IXIC',]
mydata = pd.DataFrame()

current_date = date.today()
new_date = current_date - timedelta(days=3288)

for t in tickers:
	mydata[t] = wb.DataReader(t, data_source='yahoo', start=( new_date))['Adj Close']

mydata = (mydata/mydata.iloc[0])*100

# Parse data as datetime

comp_index = []
for x in mydata.index:
    z = x.strftime("%Y-%m-%d")
    comp_index.append(z)
```

```python
fig = go.Figure()
count = -1
for t in tickers:
    count = count + 1
    fig.add_trace(go.Scatter(x=comp_index, y=mydata[t],
                    mode='lines',
                    name=names[count]))
       
fig.update_layout(title_text="<i>Credit Card Financial Services Comparables Stock Price</i> <br><b>March 2011 - March 2020</b> ",
                  xaxis_title="Year",
                  yaxis_title="Share Price in USD")


fig.show()
```

```python
# 2 Years
names = ['MasterCard', 'VISA', 'PayPal','American Express', 'Nasdaq']
tickers = ['MA', 'V','PYPL','AXP', '^IXIC']
m_mydata = pd.DataFrame()

current_date = date.today()
new_date = current_date - timedelta(days=365)

for t in tickers:
	m_mydata[t] = wb.DataReader(t, data_source='yahoo', start=new_date)['Adj Close']

m_mydata = (m_mydata/m_mydata.iloc[0])*100

# Parse data as datetime

m_comp_index = []
for x in m_mydata.index:
    f = x.strftime("%Y-%m-%d")
    m_comp_index.append(f)
```
